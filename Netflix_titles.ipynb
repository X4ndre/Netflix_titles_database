# Clening Netflix titles database
The database was taken from kaggle for more references visit: https://www.kaggle.com/datasets/rahulvyasm/netflix-movies-and-tv-shows
<br>In this project the objective is to clean any posible interference of the database for future machine learning and data science uses

```python
import pandas as pd
```

Encoding allows us to tell Pandas with what rules we want to read the file. If a value is not provided, the default encoding will be "UTC-8", which due to certain characters in this database will not be possible to decode, so in instead we use "ISO-8859-1"

```python
df = pd.read_csv("datasets/netflix_titles.csv", encoding='ISO-8859-1')
```

We search for null values within the database that may affect the database or that simply do not work for the database. An example of these are the columns "Unnamed: 12" to "Unnamed: 25" which at first glance appear to have no relevant value for the database, since they don't even have a meaningful name.

```python
df.isna().any()
```

We use the function "df.info()" to confirm the theory mentioned above and we can observe that indeed the columns "Unnamed: 12" to "Unnamed: 25" are not significant and can even be considered as errors since in these columns all values are null. This can be seen through two points, the first is "RangeIndex" which tells us the number of total data that can be in a column, in this case 8809. The second point is the "Non-Null Count" column which As its name indicates, it shows the number of data per column that are not null, which means they have an assigned value, and we can see that in the case of the "Unnamed" columns, their counter is 0, indicating that all their values are null.

```python
df.info()
```

```python
df
```

After the previous analysis we made the decision to eliminate the "Unnamed" columns

```python
def drop_all_null_columns(df):
    for i in df:
        if not df[i].notnull().any():
            print(f"No data assigned to row in columns {i}")
            df.drop([i], axis=1, inplace=True)
    return df.copy()

df = drop_all_null_columns(df)
```

We show the changes and analyze the rest of the database

```python
df.head(10)
```

In this case we observe two things, the first is that the "show_id column" is only an enumeration of the list, an enumeration that Panda already does by default and that in a case of machine learning or data science it does not represent any value added to the learning of the model, for so we can remove the column. The second thing that we can observe and that is complemented by the "df.info()" function is that there are columns that, although most of their rows have value, there are some that have null values, which in a case of machine learning or data science at the moment of training a model represents a problem since the model won't know how to interpret it. 

For the second case there are two solutions. The first is that instead of deleting the column we delete the rows with null values. The second solution is to create a variable that represents the null value, in the event that the column with null values only contains numbers, the average of each respective column where data is missing could be put in the missing rows, but in this case all the columns "object" type columns so in this case the null value can be represented with a string "0" or any other identifier.

In this case we will decide to add to each row an identifier string for null values, in this case we will choose "Void", however the identifier can be anything. In addition to eliminating the "show_id" column as we mentioned at the beginning.

```python
df.drop(["show_id"], axis=1, inplace=True)
df.head(10)
```

```python
def fill_any_null(df):
    for i in df:
        if df[i].isna().any():
            print(f"There are null values in column {i}")
            df[i] = df[i].fillna("Void")
    return df.copy()

df = fill_any_null(df)
        
```

```python
df
```

We observe that now all the columns already have a value assigned so it becomes easier to work with the database

```python
df.isna().any()
```

Now we can save this new clean database for future occasions in files such as: "csv", "html", "json", "sql", etc. In this case we will save it as "csv"

```python
df.to_csv("datasets/new_netflix_titles.csv")
```
